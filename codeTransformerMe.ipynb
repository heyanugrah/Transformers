```python
word = "anugrah is learning"
```

```python
vocab = ["<pad>","anugrah","is","learning","[SEP]","<unk>","<CLS>"];
```

```python

vocab_dict={}

#creating vocabulory
for index,vocab_item in enumerate(vocab):
  vocab_dict[index] = vocab_item
```

```python
""
print(vocab_dict)
```

```python
input_text = "anugrah is learning okay.oo"
```

```python
#Tokenize and Add CLS,PAD and SEP token with input

tokenList = []

tokenList.append("[CLS]")

for word in input_text.split():

    # Check for sentence ending with '.'
    if word.endswith("."):

        clean_word = word[:-1]

        if clean_word in vocab:
            tokenList.append(clean_word)
        else:
            tokenList.append("<unk>")

        tokenList.append("[SEP]")

    else:

        if word in vocab:
            tokenList.append(word)
        else:
            tokenList.append("<unk>")

# Add final separator if not already present
if tokenList[-1] != "[SEP]":
    tokenList.append("[SEP]")

print(tokenList)
```

```python
tokenIDList = []
print(tokenList)
```

```python
vectorList = []

#converting wordList to vectorList
for word in tokenList:
    for key, value in vocab_dict.items():
        if value == word:
            #print(word, "->", key)
            vectorList.append(key)
            tokenIDList.append(value)
```

```python
print(vectorList)
print(tokenIDList)
```

```python
max_seq_length = 10 #total of 10 words in a sentence

# Adding padding zero to make sequence length equal
while len(vectorList) < max_seq_length:
    vectorList.append(0)

vectorList
```

```python
#creating attention mask
attentionMaskList = []

attentionMaskList = [1 if token != 0 else 0 for token in vectorList]
print(attentionMaskList)
```

```python
#positional encoding
```

```python

'''
Formula:

PE(pos, 2i)   = sin(pos / 10000^(2i/d_model))
    PE(pos, 2i+1) = cos(pos / 10000^(2i/d_model))
'''
import math

pev = torch.zeros(max_seq_length, dim) #position encoding vector
pev
```

```python
pev.shape
```

```python
#position
position = torch.arange(0,10)
position
```

```python
position.shape
```

```python
#add the shape to make matrix addition
position = position.unsqueeze(1)
```

```python
position.shape
```

```python
print(pev)
```

```python
div_term = torch.exp(
    torch.arange(0, dim, 2)
    * (-math.log(10000.0) / dim)
)

#controlling oscillation speed of sin and cos waves
print(div_term)
```

```python
pev[:, 0::2] = torch.sin(position * div_term) #even apply, start from 0 to end , every second number
pev[:, 1::2] = torch.cos(position * div_term) #odd apply, start from 0 to end , every second number
```

```python
print(pev)
```

```python
combined_embedings = pev + embedings #CE = PE + Embbedings
```

```python
#Encoder
import torch.nn.functional as F

class Encoder(nn.Module):

  def __init__(self):
      super().__init__()
      self.dim = dim
      self.Wq = nn.Linear(dim,dim) #weightedQuery
      self.Wk = nn.Linear(dim,dim) #weightKey
      self.Wv = nn.Linear(dim,dim) #weightedValue

      #normalization 1
      self.norm1 = nn.LayerNorm(dim)

      #feed forward nueral network
      self.ffn = nn.Sequential(
           nn.Linear(dim,dim * 4),
           nn.ReLU(),
           nn.Linear(dim * 4,dim)
      )

      #normalization 2
      self.norm2 = nn.LayerNorm(dim)

  def selfAttention(self, x):
        # K and Q are already the results of the linear projection
        K = self.Wk(x)
        Q = self.Wq(x)
        V = self.Wv(x)

        # Use @ for matrix multiplication and .transpose() for the Key
        scores = Q @ K.transpose(-2, -1)

        print('key shape : ',K.shape)
        return scores,K,V

  def forward(self,combined_embedings):

    #calculating self attention
    scores,K,V = self.selfAttention(combined_embedings)

    #scaled self attention
    scaled_score = scores / math.sqrt(self.dim) #dimension of key

    #applying softmax to context_vector
    context_vector = F.softmax(scaled_score, dim=-1) @ V

    #adding residual connection 1 (context_vector + combined_embedings)
    residual_vector_1 = context_vector + combined_embedings

    #normalization of residual vector
    normalized_vector_1 = self.norm1(residual_vector_1)

    print('context scores: ',scores.shape)
    print('scores: ',scores.shape)
    print('scaled scores: ',scores.shape)

    #Feeding to Feed Forward nw
    ffn_output = self.ffn(normalized_vector_1)

    #residual connection 2 (normalized_vector + ffn_output)
    residual_vector_2 = normalized_vector_1 + ffn_output

    #normalization 2
    normalized_vector_2 = self.norm2(residual_vector_2)

    return normalized_vector_2,K,V
```

$$\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V$$

```python
encoderLayer = Encoder()
encoder_output,K,V = encoderLayer(combined_embedings)

print("encoder_output :" ,encoder_output)
print("key vector to decoder ", K)
print("Value vector to decoder ", V)
```
