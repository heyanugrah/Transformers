```python
input_text = "anugrah is learning okay.oo"
```

```python
#Tokenize and Add CLS,PAD and SEP token with input

tokenList = []

tokenList.append("[CLS]")

for word in input_text.split():

    # Check for sentence ending with '.'
    if word.endswith("."):

        clean_word = word[:-1]

        if clean_word in vocab:
            tokenList.append(clean_word)
        else:
            tokenList.append("<unk>")

        tokenList.append("[SEP]")

    else:

        if word in vocab:
            tokenList.append(word)
        else:
            tokenList.append("<unk>")

# Add final separator if not already present
if tokenList[-1] != "[SEP]":
    tokenList.append("[SEP]")

print(tokenList)
```

```python
#Encoder
import torch.nn.functional as F

class Encoder(nn.Module):

  def __init__(self):
      super().__init__()
      self.dim = dim
      self.Wq = nn.Linear(dim,dim) #weightedQuery
      self.Wk = nn.Linear(dim,dim) #weightKey
      self.Wv = nn.Linear(dim,dim) #weightedValue

      #normalization 1
      self.norm1 = nn.LayerNorm(dim)

      #feed forward nueral network
      self.ffn = nn.Sequential(
           nn.Linear(dim,dim * 4),
           nn.ReLU(),
           nn.Linear(dim * 4,dim)
      )

      #normalization 2
      self.norm2 = nn.LayerNorm(dim)

  def selfAttention(self, x):
        # K and Q are already the results of the linear projection
        K = self.Wk(x)
        Q = self.Wq(x)
        V = self.Wv(x)

        # Use @ for matrix multiplication and .transpose() for the Key
        scores = Q @ K.transpose(-2, -1)

        print('key shape : ',K.shape)
        return scores,K,V

  def forward(self,combined_embedings,attention_mask):

    #calculating self attention
    scores,K,V = self.selfAttention(combined_embedings)

    #scaled self attention
    scaled_score = scores / math.sqrt(self.dim) #dimension of key

    scaled_score = scaled_score.masked_fill(
        attention_mask == 0,
        -1e9
    )

    #applying softmax to scaled_score
    attention_weights = F.softmax(
        scaled_score,
        dim=-1) #-1 for each row apply softmax to all columns

    print('softmax applied :' , attention_weights)
    #creating context vecto
    context_vector = attention_weights @ V

    #adding residual connection 1 (context_vector + combined_embedings)
    residual_vector_1 = context_vector + combined_embedings

    #normalization of residual vector
    normalized_vector_1 = self.norm1(residual_vector_1)

    print('context scores: ',scores.shape)
    print('scores: ',scores.shape)
    print('scaled scores: ',scores.shape)

    #Feeding to Feed Forward Nueral Network
    ffn_output = self.ffn(normalized_vector_1)
    print('ffn_output: ',ffn_output)

    #residual connection 2 (normalized_vector + ffn_output)
    residual_vector_2 = normalized_vector_1 + ffn_output

    #normalization 2
    normalized_vector_2 = self.norm2(residual_vector_2)

    return normalized_vector_2,K,V
```

$$\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V$$

Multihead - Encoder


```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import math


class MultiHeadEncoder(nn.Module):

    def __init__(self, dim, no_of_heads,dropout):
        super().__init__()

        # ✅ CORRECT: Embedding dimension must be divisible by number of heads
        assert dim % no_of_heads == 0

        self.dim = dim
        self.no_of_heads = no_of_heads
        self.head_vector_dim = dim // no_of_heads

        print("head_vector_dim:", self.head_vector_dim)

        # Query, Key, Value projections
        self.Wq = nn.Linear(dim, dim)
        self.Wk = nn.Linear(dim, dim)
        self.Wv = nn.Linear(dim, dim)

        # Output projection
        self.Wo = nn.Linear(dim, dim)

        # Layer Normalization 1
        self.norm1 = nn.LayerNorm(dim)

        # Feed Forward Network
        self.ffn = nn.Sequential(
            nn.Linear(dim, dim * 4),
            nn.ReLU(),
            nn.Linear(dim * 4, dim)
        )

        # Layer Normalization 2
        self.norm2 = nn.LayerNorm(dim)

        #dropout
        self.dropout = nn.Dropout(dropout)

    def selfAttentionMultiHead(self, x):

        batch_size, seq_len, _ = x.shape

        print('x shape in self-attention function', x.shape)

        K = self.Wk(x)
        Q = self.Wq(x)
        V = self.Wv(x)

        print("K shape before :", K.shape)
        print("Q shape before :", Q.shape)
        print("V shape before :", V.shape)

        # ✅ CORRECT: Split embedding dimension into heads
        K = K.view(
            batch_size,
            seq_len,
            self.no_of_heads,
            self.head_vector_dim
        )

        Q = Q.view(
            batch_size,
            seq_len,
            self.no_of_heads,
            self.head_vector_dim
        )

        V = V.view(
            batch_size,
            seq_len,
            self.no_of_heads,
            self.head_vector_dim
        )

        # ✅ CORRECT: Move head dimension to front
        K = K.transpose(1, 2) #e([10, 10]) --> [2,10,5]
        Q = Q.transpose(1, 2)
        V = V.transpose(1, 2)

        print("K shape After :", K.shape)
        print("Q shape After :", Q.shape)
        print("V shape After :", V.shape)

        # ✅ CORRECT: Attention score calculation
        scores = Q @ K.transpose(-2, -1)

        return scores, K, V

    def forward(self, combined_embeddings, attention_mask):

        # Multi-head attention
        scores, K, V = self.selfAttentionMultiHead(
            combined_embeddings
        )

        # ✅ CORRECT: Scale by sqrt(head_dim)
        scaled_scores = scores / math.sqrt(
            self.head_vector_dim
        )

        # Apply mask
        scaled_scores = scaled_scores.masked_fill(
            attention_mask == 0,
            -1e9
        )

        print("scaled score:",
              scaled_scores)

        # Softmax
        attention_weights = F.softmax(
            scaled_scores,
            dim=-1 #-1 for each row apply softmax to all columns
        )
        print("attention_weights shape:",
              attention_weights.shape)


        #Applying droupout on attention-weights
        dropout_output_1 = self.dropout(
            attention_weights
            )

        print("droupout shape:",
              dropout_output_1.shape)


        # ✅ CORRECT: Context Vector
        context_vector = dropout_output_1 @ V

        print("context_vector shape before concat:",
              context_vector.shape)

        # ✅ CORRECT: Concatenate heads
        context_vector = context_vector.transpose(
            0,
            1
        )

        context_vector = context_vector.contiguous().view(
            batch_size,
            combined_embeddings.shape[1],
            self.dim
        )

        print("context_vector shape after concat:",
              context_vector.shape)

        # Output projection
        Wo = self.Wo(context_vector)

        print("Output Projection (Wo) shape :",
              Wo.shape)

        # ✅ CORRECT: Residual Connection 1
        residual_vector_1 = Wo + combined_embeddings

        # ✅ CORRECT: LayerNorm 1
        normalized_vector_1 = self.norm1(
            residual_vector_1
        )

        # Feed Forward Network
        ffn_output = self.ffn(
            normalized_vector_1
        )

        #dropout after FFN
        dropout_output_2 = self.dropout(
            ffn_output
        )

        print("droupout2 shape:",
              dropout_output_2.shape)

        # ✅ CORRECT: Residual Connection 2
        residual_vector_2 = (
            normalized_vector_1 +
            dropout_output_2
        )

        # ✅ CORRECT: LayerNorm 2
        normalized_vector_2 = self.norm2(
            residual_vector_2
        )


        return normalized_vector_2, K, V
```

```python
class TransformerEncoderStack(nn.Module):

  def __init__(self,embedding_dim,num_heads,dropout,layers):
    super().__init__()

    self.encoderLayers = nn.ModuleList([
               MultiHeadEncoder(
                    embedding_dim,
                    num_heads,
                    dropout
                )
           for _ in range(layers)
    ])

  def forward(self,combined_embedings,attention_mask):
    running_var = combined_embedings

    for encoder in self.encoderLayers:
      output, K, V = encoder(
      running_var,
      attention_mask
      )
      running_var = output

    return running_var, K, V
```
