```python
#Encoder
import torch.nn.functional as F

class Encoder(nn.Module):

  def __init__(self):
      super().__init__()
      self.dim = dim
      self.Wq = nn.Linear(dim,dim) #weightedQuery
      self.Wk = nn.Linear(dim,dim) #weightKey
      self.Wv = nn.Linear(dim,dim) #weightedValue

      #normalization 1
      self.norm1 = nn.LayerNorm(dim)

      #feed forward nueral network
      self.ffn = nn.Sequential(
           nn.Linear(dim,dim * 4),
           nn.ReLU(),
           nn.Linear(dim * 4,dim)
      )

      #normalization 2
      self.norm2 = nn.LayerNorm(dim)

  def selfAttention(self, x):
        # K and Q are already the results of the linear projection
        K = self.Wk(x)
        Q = self.Wq(x)
        V = self.Wv(x)

        # Use @ for matrix multiplication and .transpose() for the Key
        scores = Q @ K.transpose(-2, -1)

        print('key shape : ',K.shape)
        return scores,K,V

  def forward(self,combined_embedings,attention_mask):

    #calculating self attention
    scores,K,V = self.selfAttention(combined_embedings)

    #scaled self attention
    scaled_score = scores / math.sqrt(self.dim) #dimension of key

    scaled_score = scaled_score.masked_fill(
        attention_mask == 0,
        -1e9
    )

    #applying softmax to scaled_score
    attention_weights = F.softmax(
        scaled_score,
        dim=-1) #-1 for each row apply softmax to all columns

    print('softmax applied :' , attention_weights)
    #creating context vecto
    context_vector = attention_weights @ V

    #adding residual connection 1 (context_vector + combined_embedings)
    residual_vector_1 = context_vector + combined_embedings

    #normalization of residual vector
    normalized_vector_1 = self.norm1(residual_vector_1)

    print('context scores: ',scores.shape)
    print('scores: ',scores.shape)
    print('scaled scores: ',scores.shape)

    #Feeding to Feed Forward Nueral Network
    ffn_output = self.ffn(normalized_vector_1)
    print('ffn_output: ',ffn_output)

    #residual connection 2 (normalized_vector + ffn_output)
    residual_vector_2 = normalized_vector_1 + ffn_output

    #normalization 2
    normalized_vector_2 = self.norm2(residual_vector_2)

    return normalized_vector_2,K,V
```

$$\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V$$

Multihead - Encoder


```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import math


class MultiHeadEncoder(nn.Module):

    def __init__(self, dim, no_of_heads,dropout):
        super().__init__()

        # ✅ CORRECT: Embedding dimension must be divisible by number of heads
        assert dim % no_of_heads == 0

        self.dim = dim
        self.no_of_heads = no_of_heads
        self.head_vector_dim = dim // no_of_heads

        print("head_vector_dim:", self.head_vector_dim)

        # Query, Key, Value projections
        self.Wq = nn.Linear(dim, dim)
        self.Wk = nn.Linear(dim, dim)
        self.Wv = nn.Linear(dim, dim)

        # Output projection
        self.Wo = nn.Linear(dim, dim)

        # Layer Normalization 1
        self.norm1 = nn.LayerNorm(dim)

        # Feed Forward Network
        self.ffn = nn.Sequential(
            nn.Linear(dim, dim * 4),
            #nn.ReLU(),
            nn.GELU(),
            nn.Linear(dim * 4, dim)
        )

        # Layer Normalization 2
        self.norm2 = nn.LayerNorm(dim)

        #dropout
        self.dropout = nn.Dropout(dropout)

    def selfAttentionMultiHead(self, x):

        batch_size, seq_len, _ = x.shape

        print('x shape in self-attention function', x.shape)

        K = self.Wk(x)
        Q = self.Wq(x)
        V = self.Wv(x)

        print("K shape before :", K.shape)
        print("Q shape before :", Q.shape)
        print("V shape before :", V.shape)

        # ✅ CORRECT: Split embedding dimension into heads
        K = K.view(
            batch_size,
            seq_len,
            self.no_of_heads,
            self.head_vector_dim
        )

        Q = Q.view(
            batch_size,
            seq_len,
            self.no_of_heads,
            self.head_vector_dim
        )

        V = V.view(
            batch_size,
            seq_len,
            self.no_of_heads,
            self.head_vector_dim
        )

        # ✅ CORRECT: Move head dimension to front
        K = K.transpose(1, 2) #e([10, 10]) --> [2,10,5]
        Q = Q.transpose(1, 2)
        V = V.transpose(1, 2)

        print("K shape After :", K.shape)
        print("Q shape After :", Q.shape)
        print("V shape After :", V.shape)

        # ✅ CORRECT: Attention score calculation
        scores = Q @ K.transpose(-2, -1)

        return scores, K, V

    def forward(self, combined_embeddings, attention_mask):

        # Reshape attention_mask for broadcasting with scaled_scores
        # scaled_scores shape: [batch_size, num_heads, seq_len, seq_len]
        # attention_mask needs to be: [batch_size, 1, 1, seq_len]
        if attention_mask.dim() == 2: # Original shape: [batch_size, seq_len]
             attention_mask_reshaped = attention_mask.unsqueeze(1).unsqueeze(2) # Target shape: [batch_size, 1, 1, seq_len]
        else:
             attention_mask_reshaped = attention_mask # Use as is if already reshaped or has different dimensions

        # Multi-head attention
        scores, K, V = self.selfAttentionMultiHead(
            combined_embeddings
        )

        # ✅ CORRECT: Scale by sqrt(head_dim)
        scaled_scores = scores / math.sqrt(
            self.head_vector_dim
        )

        # Apply mask
        scaled_scores = scaled_scores.masked_fill(
            attention_mask_reshaped == 0,
            -1e9
        )

        print("scaled score:",
              scaled_scores)

        # Softmax
        attention_weights = F.softmax(
            scaled_scores,
            dim=-1 #-1 for each row apply softmax to all columns
        )
        print("attention_weights shape:",
              attention_weights.shape)


        #Applying droupout on attention-weights
        dropout_output_1 = self.dropout(
            attention_weights
            )

        print("droupout shape:",
              dropout_output_1.shape)


        # ✅ CORRECT: Context Vector
        context_vector = dropout_output_1 @ V

        print("context_vector shape before concat:",
              context_vector.shape)

        # ✅ CORRECT: Concatenate heads
        batch_size = combined_embeddings.shape[0]
        context_vector = context_vector.transpose(
            1,
            2
        )

        context_vector = context_vector.contiguous().view(
            batch_size,
            combined_embeddings.shape[1],
            self.dim
        )

        print("context_vector shape after concat:",
              context_vector.shape)

        # Output projection
        Wo = self.Wo(context_vector)

        print("Output Projection (Wo) shape :",
              Wo.shape)

        # ✅ CORRECT: Residual Connection 1
        residual_vector_1 = Wo + combined_embeddings

        # ✅ CORRECT: LayerNorm 1
        normalized_vector_1 = self.norm1(
            residual_vector_1
        )

        # Feed Forward Network
        ffn_output = self.ffn(
            normalized_vector_1
        )

        #dropout after FFN
        dropout_output_2 = self.dropout(
            ffn_output
        )

        print("droupout2 shape:",
              dropout_output_2.shape)

        # ✅ CORRECT: Residual Connection 2
        residual_vector_2 = (
            normalized_vector_1 +
            dropout_output_2
        )

        # ✅ CORRECT: LayerNorm 2
        normalized_vector_2 = self.norm2(
            residual_vector_2
        )


        return normalized_vector_2, K, V
```

```python
class TransformerEncoderStack(nn.Module):

  def __init__(self,embedding_dim,num_heads,dropout,layers):
    super().__init__()

    self.encoderLayers = nn.ModuleList([
               MultiHeadEncoder(
                    embedding_dim,
                    num_heads,
                    dropout
                )
           for _ in range(layers)
    ])

  def forward(self,combined_embedings,attention_mask):
    running_var = combined_embedings

    for encoder in self.encoderLayers:
      output, K, V = encoder(
      running_var,
      attention_mask
      )
      running_var = output

    return running_var, K, V
```

```python

import re

all_vocabulary = set()

for sentence in X:
    # 1. Convert to string and make it lowercase
    text = str(sentence).lower()

    # 2. CLEANING MAGIC: Remove anything that is NOT a letter (a-z) or a space (\s)
    # This automatically deletes numbers like '96' and punctuation like ')'
    clean_text = re.sub(r'[^a-z\s]', '', text)

    # 3. Split the cleaned text into words
    words = clean_text.split()

    # 4. Add the pure words to our set
    all_vocabulary.update(words)

print("Unique pure words (vocab):", len(all_vocabulary))

# Let's peek at a few of the words to prove the numbers are gone!
print(list(all_vocabulary)[:20])

# 1. Create a list of your special tokens
special_tokens = ["<PAD>", "<UNK>", "<CLS>", "<SEP>"]

# 2. Map the special tokens to IDs 0, 1, 2, 3
word_to_id = {token: idx for idx, token in enumerate(special_tokens)}

# 3. Add your clean words to the dictionary, starting at ID 4
for idx, word in enumerate(all_vocabulary):
    word_to_id[word] = idx + len(special_tokens)

print(f"Success! Your dictionary has {len(word_to_id)} total tokens.")
print("Special Token IDs:", {k: word_to_id[k] for k in special_tokens})
```

```python
import re
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# ==========================================
# 1. DATA PREPROCESSING
# ==========================================
all_token_ids = []

for sentence in X:
    clean_text = re.sub(r'[^a-z\s]', '', str(sentence).lower())
    words = clean_text.split()

    unk_id = vocab_dict["<UNK>"]
    word_ids = [vocab_dict.get(w, unk_id) for w in words]

    token_ids = [vocab_dict["<CLS>"]] + word_ids + [vocab_dict["<SEP>"]]

    if len(token_ids) > seq_len:
        token_ids = token_ids[:seq_len]
        token_ids[-1] = vocab_dict["<SEP>"]
    else:
        pad_id = vocab_dict["<PAD>"]
        token_ids = token_ids + [pad_id] * (seq_len - len(token_ids))

    all_token_ids.append(token_ids)

X_tensor = torch.tensor(all_token_ids, dtype=torch.long)

# Create attention mask
pad_id = vocab_dict["<PAD>"]
attention_mask = (X_tensor != pad_id).long()

# Convert labels to tensor
y_tensor = torch.tensor(y.values, dtype=torch.long)

# Set batch size for DataLoader
batch_size = 10 # Assuming a reasonable batch size, adjust if needed

# Create TensorDataset
dataset = TensorDataset(X_tensor, attention_mask, y_tensor)

# Create DataLoader
train_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
```

```python
import torch.nn as nn

# Define a Transformer-based model for sentiment classification
class SentimentTransformer(nn.Module):

    # Constructor: runs once when model object is created
    def __init__(
        self,
        vocab_size,      # Total number of words/tokens in vocabulary
        embedding_dim,   # Size of embedding vector for each token
        num_heads,       # Number of attention heads in multi-head attention
        dropout,         # Dropout probability
        layers,          # Number of encoder layers
        num_classes      # Number of output classes (e.g. Positive/Negative)
    ):

        # Initialize parent nn.Module class
        super(SentimentTransformer, self).__init__()

        # Embedding layer
        # Converts token IDs into dense vectors
        # Shape:
        # Input  -> (batch_size, seq_len)
        # Output -> (batch_size, seq_len, embedding_dim)
        self.embedding = nn.Embedding(
            num_embeddings=vocab_size,
            embedding_dim=embedding_dim
        )

        # Transformer Encoder Stack
        # Learns contextual relationships between words
        self.encoder = TransformerEncoderStack(
            embedding_dim=embedding_dim,
            num_heads=num_heads,
            dropout=dropout,
            layers=layers
        )

        # Final classification layer
        # Converts sentence representation into class scores
        # Shape:
        # Input  -> (batch_size, embedding_dim)
        # Output -> (batch_size, num_classes)
        self.classifier = nn.Linear(
            embedding_dim,
            num_classes
        )

    # Forward pass
    # Executes when model(input_ids, attention_mask) is called
    def forward(self, input_ids, attention_mask):

        # Convert token IDs into embeddings
        #
        # Example:
        # input_ids = [[4,7,9]]
        #
        # Output:
        # [
        #   [0.1,0.5,...],
        #   [0.7,0.2,...],
        #   [0.3,0.9,...]
        # ]
        embeddings = self.embedding(input_ids)

        # Pass embeddings through Transformer Encoder
        #
        # Encoder returns:
        # enc_out          -> contextualized token vectors
        # attention_scores -> attention weights
        # other output     -> any extra returned value
        enc_out, _, _ = self.encoder(
            combined_embedings=embeddings,
            attention_mask=attention_mask
        )

        # Select first token representation
        #
        # Shape of enc_out:
        # (batch_size, seq_len, embedding_dim)
        #
        # [:,0,:] means:
        # all batches
        # first token only
        # all embedding dimensions
        #
        # Often used as CLS token representation
        print("encoder output",enc_out)
        cls_token_out = enc_out[:, 0, :]

        # Pass CLS vector through classifier
        #
        # Output:
        # logits (unnormalized class scores)
        #
        # Example:
        # [2.3, -1.1]
        logits = self.classifier(cls_token_out)

        # Return class scores
        return logits
```

```python
# ==========================================
# 3. INITIALIZATION & TRAINING
# ==========================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = SentimentTransformer(
    vocab_size=len(vocab_dict),
    embedding_dim=100,
    num_heads=2,
    dropout=0.1,
    layers=5,
    num_classes=2
).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

epochs = 5

for epoch in range(epochs):
    model.train()
    total_loss = 0
    correct_predictions = 0
    total_samples = 0

    for batch_X, batch_mask, batch_y in train_loader:

        batch_X = batch_X.to(device)
        batch_mask = batch_mask.to(device)
        batch_y = batch_y.long().to(device)

        # The attention mask reshaping is now handled inside the MultiHeadEncoder.
        # The original batch_mask (shape [batch_size, seq_len]) is passed directly.

        optimizer.zero_grad()

        logits = model(input_ids=batch_X, attention_mask=batch_mask)
        loss = criterion(logits, batch_y)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        predictions = torch.argmax(logits, dim=1)
        correct_predictions += (predictions == batch_y).sum().item()
        total_samples += batch_y.size(0)

    avg_loss = total_loss / len(train_loader)
    accuracy = (correct_predictions / total_samples) * 100
    print(f"Epoch {epoch+1}/{epochs} | Loss: {avg_loss:.4f} | Accuracy: {accuracy:.2f}%")

print("Training Complete! 🚀")



```

```python
# ==========================================
# 4. PREDICTION FOR A NEW SENTENCE
# ==========================================

def predict_sentiment(sentence: str, model, vocab_dict, seq_len, device):
    model.eval() # Set model to evaluation mode

    # 1. Preprocessing (similar to training data)
    clean_text = re.sub(r'[^a-z\s]', '', str(sentence).lower())
    words = clean_text.split()

    unk_id = vocab_dict["<UNK>"]
    # Convert words to IDs, add CLS and SEP tokens
    token_ids = [vocab_dict["<CLS>"]] + [vocab_dict.get(w, unk_id) for w in words] + [vocab_dict["<SEP>"]]

    # Apply sequence length truncation or padding
    if len(token_ids) > seq_len:
        token_ids = token_ids[:seq_len]
        token_ids[-1] = vocab_dict["<SEP>"] # Ensure SEP is at the end if truncated
    else:
        pad_id = vocab_dict["<PAD>"]
        token_ids = token_ids + [pad_id] * (seq_len - len(token_ids))

    # Convert to tensor
    input_ids = torch.tensor([token_ids], dtype=torch.long).to(device)

    # Create attention mask
    attention_mask = (input_ids != vocab_dict["<PAD>"]).long().to(device)

    # 2. Make prediction
    with torch.no_grad(): # Disable gradient calculations for inference
        logits = model(input_ids=input_ids, attention_mask=attention_mask)
        probabilities = torch.softmax(logits, dim=1)
        predicted_class = torch.argmax(probabilities, dim=1).item()

    # 3. Interpret result
    sentiment_label = "Positive" if predicted_class == 1 else "Negative"
    return sentiment_label, probabilities[0][predicted_class].item()

# Example Usage:
new_sentence_1 = "This movie was absolutely fantastic, I loved every minute of it!"
new_sentence_2 = "good day."

# Assuming `vocab_dict` and `seq_len` are defined from previous cells
predicted_sentiment_1, confidence_1 = predict_sentiment(new_sentence_1, model, vocab_dict, seq_len, device)
print(f"Sentence: '{new_sentence_1}'\nPredicted Sentiment: {predicted_sentiment_1} (Confidence: {confidence_1:.2f})\n")

predicted_sentiment_2, confidence_2 = predict_sentiment(new_sentence_2, model, vocab_dict, seq_len, device)
print(f"Sentence: '{new_sentence_2}'\nPredicted Sentiment: {predicted_sentiment_2} (Confidence: {confidence_2:.2f})\n")
```
